```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import pylab as pl 
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
digits=load_digits()
```

```python
pl.gray()
pl.matshow(digits.images[14])
pl.show()
```

```python
digits.images[14]
```

```python
images_and_lables= list(zip(digits.images,digits.target))
plt.figure(figsize=(5,5))
for index ,(image ,label) in enumerate (images_and_lables[:15]):
    plt.subplot(3,5,index +1)
    plt.axis('off')
    plt.imshow(image , cmap=plt.cm.gray_r , interpolation='nearest')
    plt.title('%i' % label)
```

```python
n_samples= len(digits.images)
print(n_samples)
```

```python
X=digits.images.reshape((n_samples , -1))
y= digits.target 
```

```python
X_train ,X_test, y_train ,y_test =train_test_split(X,y)
print(X_train.shape)
print(X_test.shape)
```

```python
model_linear =svm.SVC(kernel='linear' , degree=3 ,gamma='scale')
model_linear.fit(X_train, y_train)

y_pred =model_linear.predict(X_test)
```

```python
model_linear.score(X_test ,y_test)
```

```python
model_RBF =svm.SVC(degree=3 ,gamma='scale' ,kernel='rbf')
model_RBF.fit(X_train ,y_train)

y_pred =model_RBF.predict(X_test)
model_RBF.score(X_test ,y_test)
```

```python
from sklearn.metrics import classification_report
predictions =model_linear.predict(X_test)
print (classification_report(y_test, predictions))
```
